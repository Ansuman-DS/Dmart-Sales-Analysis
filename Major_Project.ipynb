# ***Model*** **Building**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
lr= LinearRegression()
```

```python
lr1= lr.fit(X_train_sd,Y_train)
```

```python
y_pred_lr=lr1.predict(X_test_sd)
y_pred_lr
```

```python
Y_test
```

```python
scores = cross_val_score(lr, X_train, Y_train, cv=5, scoring='r2')
scores.mean()
```

```python
score_lr = r2_score(Y_test,y_pred_lr)
score_lr
```

```python
mean_absolute_error(Y_test,y_pred_lr)
```

```python
np.sqrt(mean_squared_error(Y_test,y_pred_lr))
```

```python
# Ensure Y_test is sorted in the same order as y_pred
Y_test_sorted = Y_test.sort_index()

# Define a tolerance level, for example, 10%
tolerance = 0.10

# Calculate the accuracy within the tolerance level
accurate_predictions = np.abs(y_pred_lr - Y_test_sorted) / Y_test_sorted <= tolerance
accuracy_lr = np.mean(accurate_predictions)

accuracy_lr=accuracy_lr*100
accuracy_lr
```

```python

# Accuracy data (in range 0 to 0.09)
models = ['Linear Regression', 'Random Forest', 'XGBRFRegressor']
accuracies = [accuracy_lr, accuracy_rfe, accuracy_xg]  # Replace these with your actual accuracy values

# Create a DataFrame
data = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracies
})

# Determine the best model
best_model_index = accuracies.index(max(accuracies))

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Model', y='Accuracy', data=data, palette=['blue', 'cyan', 'blue'])
bar_plot.patches[best_model_index].set_color('red')

# Add titles and labels
plt.title('Accuracy Comparison of ML Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0, 10)  # Adjust the y-axis to fit your data range

# Display the accuracy values on the bars
for i, bar in enumerate(bar_plot.patches):
    bar_plot.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.002, f'{accuracies[i]:.4f}', ha='center', va='bottom')

# Annotate the best model
plt.text(bar_plot.patches[best_model_index].get_x() + bar_plot.patches[best_model_index].get_width() / 2,
         accuracies[best_model_index] + 0.9, 'Best Model', ha='center', va='bottom', color='green', fontweight='bold')

# Show the plot
plt.show()
```
